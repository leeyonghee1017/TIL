# I. 실습 데이터

> ## 1) seaborn 'mpg' Data Set

* 자동차 연비(mpg) 데이터
  * mpg : miles per gallon
  * 1 mile : 1.6 Km
  * 1 gallon : 3.78 Liter

```python
import seaborn as sns
DF = sns.load_dataset('mpg')
```

```python
DF.info()
```

* 문자형 데이터 : 'origin'

```python
DF.head()
```

```python
type(DF.origin[0])
```

* 명목형 : 이름확인 및 빈도분석

```python
DF.origin.value_counts()
```

* 'origin' Data

```python
X = DF[['origin']]
```

```python
X[111:115]
```

> ## 2) With LabelEncoder

* 정수(Integer) 인코딩

```python
from sklearn.preprocessing import LabelEncoder
encoder1 = LabelEncoder()
LE = encoder1.fit_transform(X)
```

* 정수 인코딩 결과

```python
LE[111:115]
```

> ## 3) With OneHotEncoder

* 원-핫(One-Hot) 인코딩

```python
from sklearn.preprocessing import OneHotEncoder
encoder2 = OneHotEncoder()
OHE = encoder2.fit_transform(X)
```

* Array 변환 필요

```python
print(OHE[111:115])
```

```python
OHE.toarray()[111:115]
```
