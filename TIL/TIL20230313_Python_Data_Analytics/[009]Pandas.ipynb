> # 8) Indexing & Slicing

* Lable(Index) 변경

```python
LABEL = ['No_1', 'No_2', 'No_3', 'No_4', 'No_5', 'No_6', 'No_7', 'No_8', 'No_9', 'No_10', 'No_11', 'No_12', 'No_13', 'No_14', 'No_15', 'No_16', 'No_17']
DF1.index = LABEL
```

```python
DF1
```

> ## (1) loc(Location)

* Label 값 기준
* loc - indexing
  * .loc[행_label, 열_label]

```python
DF1.loc['No_3', 'Age']
```

* loc - Slicing
  * .loc[행_label_시작:행_label_끝, 열_label_시작:열_label_끝]
  * '행_label_시작' 이상 '행_label_끝' 이하
  * '열_label_시작' 이상 '열_label_끝' 이하

```python
DF1.loc['No_3':'No_6', 'Age':'BloodType']
```

> ## (2) iloc(Integer Location)

* Index 값 기준
* iloc - indexing

```python
DF1.iloc[2, 2]
```

* iloc - Slicing
  * .iloc[행_index_시작:행_index_끝, 열_index_시작:열_index_끝]
  * '행_index_시작' 이상 '행_index_끝' 미만
  * '열_index_시작' 이상 '열_index_끝' 미만

```python
DF1.iloc[2:5, 2:5]
```

# III. pandas - Visualization
* https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html




> # 1) 선 그래프

* kind = 'line'
* style: '-','--','-.', 's-', 'o-', '^-'
* linewidth: 숫자
* color: 16진수 RGB
* figsize: inch

```python
DF1[['Height']].plot(kind = 'line',style = '-',linewidth = 2,color = '#FF0000',grid = True,figsize = (10, 7))
```

> # 2) 막대 그래프

* kind = 'bar'
* width: 0~1
* rot: 0~360
* fontsize: 숫자

```python
DF1[['Height', 'Weight']].plot(kind = 'bar',width = 0.5,rot = 45,fontsize = 15,figsize = (10, 7))
```

* kind = 'barh'

```python
DF1[['Height', 'Weight']].plot(kind = 'barh',rot = 45,color = ['#FA5800', '#A20025'],fontsize = 15,figsize = (7, 10))
```

> # 3) 히스토그램

* kind = 'hist'
* bins: 구간 개수
* alpha: 0~1

```python
DF1['Height'].plot(kind = 'hist',bins = 5,alpha = 0.5,figsize = (10, 7))
```

> # 4) 상자 그래프

* kind = 'box'

```python
DF1['Height'].plot(kind = 'box',figsize = (7, 9))
```

> 5) 산점도

* kind = 'scatter'
* s: 숫자

```python
DF1[['Height', 'Weight']].plot(kind = 'scatter',x = 'Height',y = 'Weight',s = 50,figsize = (10, 7))
```

> 6) 파이 그래프

* kind = 'pie'

```python
DF1.BloodType.value_counts().plot(kind = 'pie',autopct = '%.1f%%',fontsize = 15,figsize = (10, 10))
```
