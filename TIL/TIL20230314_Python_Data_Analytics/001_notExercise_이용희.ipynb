# I. Data Preprocessing

```
import warnings
warnings.filterwarnings('ignore')
```

> ## 1) 'notExercise.xlsx' 파일 읽어오기

```
import pandas as pd

url = 'https://raw.githubusercontent.com/rusita-ai/pyData/master/notExercise.xlsx'
DF = pd.read_excel(url)

DF.head()
```

* DataFrame 정보

```
DF.info()
```

> ## 2) '기간' 열(Column) 삭제

* '기간'열(Column) 정보

```
DF['기간'].info()
```

* '기간'열(Column) <font color = 'red'>삭제

```
DF.drop(['기간'],axis=1,inplace=True)
```

* <font color = 'red'>삭제</font> 결과 확인

```
DF.head(5)
```

> ## 3) 0, 23~52번(지역분류) index 삭제

* '대분류' 정보

```
DF['대분류']
```

* 23번 ~ 52번 index <font color = 'red'>삭제

```
DF.drop(DF.index[23:53],axis=0,inplace=True)
```

```
DF
```

* 0번 index <font color = 'red'>삭제

```
DF.drop([0],axis=0,inplace=True)
```

```
DF
```

# II. '성별' 통계

> ## 1) '성별' 데이터 추출

```
data_1 = DF.loc[(DF.대분류 == "성별"), :]
```

```
data_1
```

> ## 2) '대분류' 열(Column) 삭제

```
data_1.drop(['대분류'],axis=1,inplace=True)
```

```
data_1
```

> ## 3) '분류'열(Column)을 index로 지정

```
data_2 = data_1.set_index("분류")
```

```
data_2
```

> ## 4) matplotlib 시각화

```
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (6, 6))

ax[0, 0].pie(data_2[data_2.columns[0]],
labels = data_2[data_2.columns[0]].index,
autopct = '%.1f%%',
explode = [0.0, 0.0])

ax[0, 1].pie(data_2[data_2.columns[1]],
labels = data_2[data_2.columns[1]].index,
autopct = '%.1f%%',
explode = [0.0, 0.0])

ax[1, 0].pie(data_2[data_2.columns[2]],
labels = data_2[data_2.columns[2]].index,
autopct = '%.1f%%',
explode = [0.0, 0.0])

ax[1, 1].pie(data_2[data_2.columns[3]],
labels = data_2[data_2.columns[3]].index,
autopct = '%.1f%%',
explode = [0.0, 0.0])

ax[0, 0].plot()
ax[0, 0].set_title(data_2.columns[0])
ax[0, 0].set_xticks([])
ax[0, 0].set_yticks([])
ax[0, 1].plot()
ax[0, 1].set_title(data_2.columns[1])
ax[0, 1].set_xticks([])
ax[0, 1].set_yticks([])
ax[1, 0].plot()
ax[1, 0].set_title(data_2.columns[2])
ax[1, 0].set_xticks([])
ax[1, 0].set_yticks([])
ax[1, 1].plot()
ax[1, 1].set_title(data_2.columns[3])
ax[1, 1].set_xticks([])
ax[1, 1].set_yticks([])
plt.show()
```
