# I. Machine( ) 정의

* numpy Package

```python
import numpy as np
```

* def Machine( )


```python
def Machine(x, w, b):
  y_hat = (w * x) + b
  return y_hat
```

* x, w, b 객체 지정

```python
x = np.array([1, 3, 5, 7, 9])
w = 2
b = 1
```

* Machine( ) 테스트

```python
Machine(x, w, b)
```

# II. Gradient( ) 정의

* def Gradient( )

```python
def Gradient(x, y, w, b):
  y_hat = Machine(x, w, b)
  dw = np.mean((y - y_hat) * (-2 * x))
  db = np.mean((y - y_hat) * (-2))
  return dw, db
```

* Gradient( ) 테스트

```python
y = np.array([2, 4, 6, 8, 10])
dw, db = Gradient(x, y, w, b)
```

```python
print('dw is ', dw)
print('db is ', db)
```

# III. Learning( ) 정의

* def Learning( )

```python
def Learning(x, y, w, b, step):
  dw, db = Gradient(x, y, w, b)
  uw = w - step * dw
  ub = b - step * db
  return uw, ub
```

* Learning( ) 테스트

```python
step = 0.05
uw, ub = Learning(x, y, w, b, step)
```

```python
print('Updated_w is ', '%.3f' % uw)
print('Updated_b is ', '%.3f' % ub)
```

# V. Loss(MSE) Visualization

* Gradient( )에 MSE 추가

```python
def Gradient(x, y, w, b):
  y_hat = Machine(x, w, b)
  dw = np.mean((y - y_hat) * (-2 * x))
  db = np.mean((y - y_hat) * (-2))
  MSE = np.mean((y - y_hat)**2)
  return dw, db, MSE
```

* Learning( )에 MSE 추가

```python
def Learning(x, y, w, b, step):
  dw, db, MSE = Gradient(x, y, w, b)
  uw = w - step * dw
  ub = b - step * db
  return uw, ub, MSE
```

* 1500번 학습 실행

```python
w = 2
b = 3
step = 0.001
mse = []
```

```python
for i in range(0, 1500):
  uw, ub, MSE = Learning(DATA.inputs, DATA.outputs, w, b, step)
  w = uw
  b = ub
  mse.append(MSE)
```

* Loss(MSE) 감소 확인

```python
mse[0:10]
```

```python
plt.plot(mse)
plt.show()
```
